# What are the top 5 most hottest years

```python
most=temp.sort_values('temp_celsius',ascending=False).head(5)
```

```python
plt.figure(figsize=(8,10))
sns.catplot(x=" Year", y="temp_celsius", hue="Month", kind="bar", palette="rocket",data=most)
```

# Take a mean of Each Year

```python
xyears=temp[' Year'].unique()
temp_data = pd.DataFrame(columns={'Year', 'Mean'})
rows=0
for x in xyears:
  temp_data.at[rows,'Year']=x
  temp_data.at[rows,'Mean']=temp[temp[' Year']==x]['temp_celsius'].mean()
  rows+=1
temp_data
```

```python
temp_data['Mean']=temp_data['Mean'].astype('float')
temp_data.plot('Year','Mean')
```

# Last Five Years Data Visualization

```python
temp_5 = temp[(temp[' Year']>=2012)]
temp_5
```

```python
xyears_temp_5=temp_5[' Year'].unique()
temp_5_data=pd.DataFrame(columns={'Year','Mean'})
rows = 0
for x in xyears_temp_5:
  temp_5_data.at [rows,'Year']=x
  temp_5_data.at [rows,'Mean']=temp_5[temp_5[' Year']==x]['temp_celsius'].mean()
  rows+=1
temp_5_data
```

```python
temp_5_data.plot('Year','Mean',color='green')
```
