# **Top 20 Oil Consumers**

```python
#convert data type from object to float
data['World Share']=data['World Share'].str.replace('%','').astype('float')
```

```python
top_country = data.nlargest(20,'World Share')
top_country
```

```python
# visualize in chart
plt.figure(figsize=(15,8))
sns.barplot(x=top_country['Country'],y=top_country['World Share'])
plt.xticks(rotation=90,fontsize=14)
plt.xlabel('Country',fontsize=18)
plt.ylabel('Percentage of consumption',fontsize=18)
plt.title('Top 20 Oil consumer countries',fontsize=20)

```

```python
plt.figure(figsize=(15,15))
world_share=data['World Share']
country=data['Country']
palette_color = sns.color_palette('bright')
plt.pie(world_share,labels=country,colors=palette_color, rotatelabels=True, autopct='%.0f%%')
plt.xticks(rotation=90)
plt.show()
```

# **Top 20 Oil Consumers per capita**

```python
top_per_capita=data.nlargest(20,'Yearly Gallons Per Capita')
top_per_capita.head()
```

```python
#visualization in barplot
plt.figure(figsize=(15,8))
sns.barplot(x='Country',y='Yearly Gallons Per Capita',data=top_per_capita)
plt.xticks(rotation=90,fontsize=14)
plt.xlabel('Country',fontsize=18)
plt.ylabel('Percentage of consumption',fontsize=18)
plt.title('Top 20 Per Capita',fontsize=20)
```

# **GDP to Gallon GDP per capita can buy correlation of top 10 Expensive countries**

```python
#convert datatype
data['GDP Per Capita ( USD )']=data['GDP Per Capita ( USD )'].str.replace(',','').astype('float')
data['Gallons GDP Per Capita Can Buy']=data['Gallons GDP Per Capita Can Buy'].str.replace(',','').astype('float')
```

```python
top_expensive = data.nlargest(10,'Price Per Liter (USD)').head()
top_expensive
```

```python
corr_exp= top_expensive['GDP Per Capita ( USD )'].corr(top_expensive['Gallons GDP Per Capita Can Buy'])
print(corr_exp)
```

# **GDP to Gallon GDP per capita can buy Pakistan**
